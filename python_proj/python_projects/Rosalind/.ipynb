```python
def alpha_combs(alphabet, n, acc='', res=[]):
    if n == 0:
        res.append(acc)
    else:
        for c in alphabet:
            
            alpha_combs(alphabet, n - 1, acc + c, res)
            #print(acc+c)
    return res
```

```python
def kmer_gen():
    small_dataset = "T A G C\n4"
    #large_dataset = open('D:/python_proj/python_projects/Rosalind/rosalind_lexf.txt').read()
    #bits = large_dataset.split()
    bits = small_dataset.split()
    alphabet = bits[:-1]
    alphabet = ['1','2','3','4']
    comb_len = int(bits[-1])
    kmer_list = list()
    for p in alpha_combs(alphabet, comb_len):
        kmer_list.append(p)
        #print(p)
    return(sorted(kmer_list))
```
